```python
# Visualize average ratings for each user using a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("Pandas Version Number: " + pd.__version__)
```

```python
# Input csv file, making first column the index
#average_ratings_df = pd.read_csv('../../../data/average_ratings/average_ratings_test.csv', index_col = 0) ##### For testing
average_ratings_df = pd.read_csv('../../../data/average_ratings/average_ratings.csv', index_col = 0)
display(average_ratings_df)
```

```python
# Truncate the DataFrame to a feasible display size
average_ratings_df = average_ratings_df.head(100)
display(average_ratings_df)
```

```python
# Set y-axis label
average_ratings_df.index.names = ['User ID']

# Change size of the heatmap as well as font size
plt.figure(figsize = (25,50))
sns.set(font_scale = 2.5)

# Create heatmap and change settings
heatmap = sns.heatmap(average_ratings_df, xticklabels = True, yticklabels = True, cmap = 'PuRd', square = True)
heatmap.set_title('Average User Ratings for Genres of Anime Series')
heatmap.set_xlabel('Genre')
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation = 0)
```
